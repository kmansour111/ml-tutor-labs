```python
# WORKED EXAMPLE — the whole course in 12 lines: load → clean → split → fit → honest score
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

df = pd.DataFrame({   # 8 synthetic patients; one missing value to clean
    "age":       [72, 34, 58, 41, 66, 29, 77, 50],
    "potassium": [6.1, 4.2, None, 4.8, 5.9, 4.0, 6.3, 4.4],
    "flag":      [1, 0, 1, 0, 1, 0, 1, 0],
})
df["potassium"] = df["potassium"].fillna(df["potassium"].median())        # Week 3 habit
Xtr, Xte, ytr, yte = train_test_split(df[["age", "potassium"]], df["flag"],
                                      test_size=0.25, random_state=0)     # Week 5 habit
model = DecisionTreeClassifier(max_depth=2, random_state=0).fit(Xtr, ytr) # Week 2 model
print("held-out accuracy:", model.score(Xte, yte))   # judged on data it NEVER saw — Week 4+ habit
```
