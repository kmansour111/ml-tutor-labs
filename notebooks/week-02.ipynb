## Worked example — Supervised learning
**Lane A · the general idea:** In supervised learning, each training row comes with the answer you want the model to learn. The model searches for patterns linking input features to that target label.

**Lane B · the same idea in pharmacy terms:** For adherence, the label might be `late_refill_next_month` and the features might include age band, refill gap history, number of chronic medications, and prior missed pickups. We are not giving advice; we are studying a prediction setup.

Below is a tiny, complete demo of this idea. Run it, read every line, change one thing, run it again.


```python
# WORKED EXAMPLE — a decision tree LEARNS an adherence rule from 6 patients
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

df = pd.DataFrame({
    "age":           [72, 34, 58, 41, 66, 29],
    "n_medications": [9,  1,  6,  2,  8,  1],
    "adherent":      [0,  1,  0,  1,  0,  1],   # target: 1 = adherent (synthetic)
})
X, y = df[["age", "n_medications"]], df["adherent"]
model = DecisionTreeClassifier(max_depth=2, random_state=0).fit(X, y)

new_patient = pd.DataFrame({"age": [70], "n_medications": [7]})
print("Predicted adherent?", model.predict(new_patient)[0])   # rule came from DATA, not from us
```
