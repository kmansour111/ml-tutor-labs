## Worked example — Retrieval before generation
**Lane A · the general idea:** RAG searches an approved corpus for relevant chunks, then conditions generation on those chunks. It is a practical way to keep answers closer to the source material.

**Lane B · the same idea in pharmacy terms:** For a grounded pharmacy assistant, retrieval ensures the response stays tied to approved documents instead of drifting into unsupported drug or workflow claims.

Below is a tiny, complete demo of this idea. Run it, read every line, change one thing, run it again.


```python
# WORKED EXAMPLE — retrieval = score documents by overlap with the question
corpus = {
    "doc1": "warfarin dosing is adjusted using INR monitoring",
    "doc2": "metformin is first line therapy for type 2 diabetes",
    "doc3": "INR above range increases bleeding risk with warfarin",
}
question = "how does INR guide warfarin dosing"

q_words = set(question.lower().split())
scores = {d: len(q_words & set(t.lower().split())) for d, t in corpus.items()}
print(sorted(scores.items(), key=lambda kv: -kv[1]))
# RAG = do this (with better scoring), then hand the TOP documents to the LLM as context.
```
