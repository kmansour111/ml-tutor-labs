## Worked example — Packaging and deployment
**Lane A · the general idea:** Packaging means putting the trained model behind a stable interface such as a Streamlit page or an API. Deployment makes the model usable outside the notebook.

**Lane B · the same idea in pharmacy terms:** A medication dashboard is the teaching example here: the model needs a small, reproducible interface so its outputs can be inspected and discussed safely.

Below is a tiny, complete demo of this idea. Run it, read every line, change one thing, run it again.


```python
# WORKED EXAMPLE — a model is just a FILE: save, reload, predict (deployment step 0)
import pickle
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

df = pd.DataFrame({"age": [72, 34, 58, 41], "n_meds": [9, 1, 6, 2], "risk": [1, 0, 1, 0]})
model = DecisionTreeClassifier(random_state=0).fit(df[["age", "n_meds"]], df["risk"])

with open("model.pkl", "wb") as f:
    pickle.dump(model, f)          # ship THIS file, not your notebook
with open("model.pkl", "rb") as f:
    served = pickle.load(f)        # what the dashboard loads at startup
print("served prediction:", served.predict(pd.DataFrame({"age": [70], "n_meds": [8]}))[0])
```
