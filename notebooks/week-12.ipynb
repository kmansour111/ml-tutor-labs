## Worked example — Hallucination and sycophancy
**Lane A · the general idea:** A model hallucinates when it produces a plausible but unsupported answer. Sycophancy happens when it tries too hard to please the user rather than staying accurate.

**Lane B · the same idea in pharmacy terms:** In a health-science setting, a hallucinated answer can sound polished while inventing a fact or citation. The safer response is to check the claim against a trusted source before sharing it.

Below is a tiny, complete demo of this idea. Run it, read every line, change one thing, run it again.


```python
# WORKED EXAMPLE — verify an AI claim against the source, MECHANICALLY
source_facts = {   # your trusted reference material (teaching excerpt)
    "metformin max daily dose": "2550 mg (immediate-release, adults)",
    "metformin renal cutoff": "contraindicated if eGFR < 30 mL/min/1.73m2",
}
ai_claim = ("metformin max daily dose", "3000 mg for all adults")   # sounds confident!

topic, claimed = ai_claim
print("AI claims  :", claimed)
print("Source says:", source_facts.get(topic, "NOT IN SOURCE — cannot verify"))
print("Match?     :", claimed == source_facts.get(topic))
# fluent ≠ true. Verification is a LOOKUP against a source, not a vibe.
```
