## Worked example — Overfitting and regularization
**Lane A · the general idea:** Overfitting happens when a model fits the training data too closely and loses generalization. Regularization methods such as dropout, augmentation, and early stopping help control that behavior.

**Lane B · the same idea in pharmacy terms:** In pill-image classification, overfitting can make the model memorize lighting or background artifacts instead of pill shape. The fix is not more confidence; it is a better training recipe.

Below is a tiny, complete demo of this idea. Run it, read every line, change one thing, run it again.


```python
# WORKED EXAMPLE — a memorizer vs a learner (overfitting in 10 lines)
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

# 200 synthetic patients with noisy labels (flip_y) — like real clinical data
X, y = make_classification(n_samples=200, n_features=8, flip_y=0.15, random_state=0)
Xtr, Xte, ytr, yte = train_test_split(X, y, random_state=0)

for depth in [None, 3]:                  # None = grow until it MEMORIZES the noise
    m = DecisionTreeClassifier(max_depth=depth, random_state=0).fit(Xtr, ytr)
    print(f"max_depth={depth}: train={m.score(Xtr, ytr):.2f}  test={m.score(Xte, yte):.2f}")
# a big train/test gap = overfitting; the depth limit is REGULARIZATION at work.
```
