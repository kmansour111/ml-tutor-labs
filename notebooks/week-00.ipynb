### f-strings — the one string tool you'll use constantly

Put `f` before the quotes; anything in `{curly braces}` gets filled in with a value.


```python
patient = "MRN-1042"
potassium = 6.1
print(f"Patient {patient}: potassium = {potassium} mEq/L")

# Formatting decimals: {value:.2f} -> 2 decimal places
crcl = 47.3333
print(f"CrCl = {crcl:.1f} mL/min")
```

### The trap that bites every beginner

Numbers-as-text look identical when printed but behave completely differently. Data from files ARRIVES as text — you convert it once at the boundary, then compute with numbers.


```python
text_value = "6.1"        # a string that LOOKS like a number
real_value = 6.1          # an actual float

print(text_value + text_value)    # "6.16.1"  — glued text!
print(real_value + real_value)    # 12.2      — arithmetic

converted = float(text_value)     # deliberate conversion
print(converted + converted)      # 12.2
print(type(text_value), type(real_value))   # check any value's type with type()
```

## 4 · Booleans & comparisons — True / False

Every clinical rule reduces to expressions that come out `True` or `False`.


```python
potassium = 6.1
print(potassium > 5.0)     # True
print(potassium == 5.0)    # False   (== asks equality; a single = ASSIGNS!)
print(potassium != 5.0)    # True    (not equal)
print(3.5 <= potassium <= 5.0)   # False — Python allows chained ranges
```

```python
# Combine conditions with:  and, or, not
on_ace_inhibitor = True
print(potassium > 5.0 and on_ace_inhibitor)   # both must be True
print(potassium > 5.0 or on_ace_inhibitor)    # at least one True
print(not on_ace_inhibitor)                   # flips it
```

## 5 · Decisions — if / elif / else

Read it out loud and it says what it does. Two things matter:
1. The colon `:` at the end of each condition line.
2. **Indentation** (4 spaces) marks which lines belong to which branch. The first true branch wins; the rest are skipped.


```python
potassium = 6.1

if potassium >= 6.0:
    print("CRITICAL — page the team")
elif potassium > 5.0:
    print("HIGH — flag for review")
elif potassium < 3.5:
    print("LOW — flag for review")
else:
    print("within range")

# Try 4.2, then 3.1, then 6.5 — re-run each time.
```

**Boundary values are clinical decisions.** `>= 6.0` vs `> 6.0` decides what happens at *exactly* 6.0. Write the rule to match the protocol, then test the exact boundary — a habit we drill all semester.

## 10 · Functions — name a recipe, reuse it

`def` defines a function: inputs in parentheses (**parameters**), result handed back with **return**. Functions are how code stops being a script and starts being a toolkit.


```python
def creatinine_clearance(age, weight_kg, scr, female):
    """Cockcroft-Gault estimate (teaching version)."""
    crcl = ((140 - age) * weight_kg) / (72 * scr)
    if female:
        crcl = crcl * 0.85
    return crcl

# Call it as many times as you like, with different inputs:
print(f"{creatinine_clearance(62, 70, 1.1, False):.1f} mL/min")
print(f"{creatinine_clearance(80, 55, 1.4, True):.1f} mL/min")
```

```python
def is_out_of_range(value, low, high):
    return value < low or value > high      # returns True or False directly

print(is_out_of_range(6.1, 3.5, 5.0))   # True
print(is_out_of_range(4.2, 3.5, 5.0))   # False
```

**Note:** `return` hands the value back so other code can use it; `print` only *shows* it. Beginner bug #1 is printing inside a function when you needed to return.

## 15 · Data quality — the habit that saves careers

Models are only as good as the data. In health data the classic defects are: **missing values**, **impossible values** (age 212, potassium 61), **unit mix-ups** (mg vs mcg), **duplicates**, and **strings pretending to be numbers**. You already have every tool needed to CHECK for these — that's the point of this section.


```python
# A messy mini-dataset — exactly the kind of thing a CSV export gives you:
raw = [
    {"patient_id": "MRN-1042", "analyte": "potassium", "value": 6.1,   "unit": "mEq/L"},
    {"patient_id": "MRN-2117", "analyte": "potassium", "value": "4.2", "unit": "mEq/L"},  # string!
    {"patient_id": "MRN-3306", "analyte": "potassium", "value": None,  "unit": "mEq/L"},  # missing
    {"patient_id": "MRN-4478", "analyte": "potassium", "value": 61.0,  "unit": "mEq/L"},  # impossible — unit error?
    {"patient_id": "MRN-1042", "analyte": "potassium", "value": 6.1,   "unit": "mEq/L"},  # duplicate
]

# QC sweep with tools you already know:
for i, rec in enumerate(raw):
    v = rec["value"]
    if v is None:
        print(f"row {i}: MISSING value")
    elif type(v) is str:
        print(f"row {i}: value is TEXT ('{v}') — convert with float() at the boundary")
    elif v > 10:
        print(f"row {i}: {v} is physiologically impossible — suspect unit error")
```

```python
# Duplicates: same patient + analyte + value appearing twice
seen = []
for rec in raw:
    key = (rec["patient_id"], rec["analyte"], rec["value"])
    if key in seen:
        print("DUPLICATE:", key)
    seen.append(key)
```

**The Week-0 data-quality creed** (this becomes CO4 all semester):
1. Never trust a dataset you haven't profiled — check types, missing, ranges, duplicates FIRST.
2. Convert text→numbers once, at the boundary, deliberately.
3. Impossible values are usually *unit* or *entry* errors — investigate, don't delete blindly.
4. Write your checks as code (like above) so they run every time, not just the day you remembered.


### 15b · The systematic QC checklist — run it on EVERY dataset

Ad-hoc checks miss things. Professionals run the same six checks, in the same order, every time:

| # | Check | Question | Tool |
|---|-------|----------|------|
| 1 | **Shape** | How many rows/columns? Is that what I expected? | `df.shape` |
| 2 | **Types** | Are numbers numeric? Dates dates? | `df.dtypes` |
| 3 | **Missing** | How much is missing, per column? Random or patterned? | `df.isna().sum()` |
| 4 | **Ranges** | Any impossible values? (age 212, K 61) | `df.describe()`, domain limits |
| 5 | **Duplicates** | Exact repeats? Same patient+time twice? | `df.duplicated()` |
| 6 | **Units & categories** | mg vs mcg? "M"/"male"/"MALE" as three categories? | `df["col"].unique()` |


```python
import pandas as pd
import numpy as np

# A realistically messy extract — every classic defect planted:
labs = pd.DataFrame([
    {"patient_id": "MRN-1042", "sex": "F",      "age": 62,  "analyte": "potassium", "value": 6.1,    "unit": "mEq/L", "collected": "2026-07-01 07:15"},
    {"patient_id": "MRN-2117", "sex": "female", "age": 80,  "analyte": "potassium", "value": "4.2",  "unit": "mEq/L", "collected": "2026-07-01 07:20"},
    {"patient_id": "MRN-3306", "sex": "F",      "age": 45,  "analyte": "potassium", "value": None,   "unit": "mEq/L", "collected": "2026-07-01 07:31"},
    {"patient_id": "MRN-4478", "sex": "M",      "age": 212, "analyte": "potassium", "value": 4.9,    "unit": "mEq/L", "collected": "2026-07-01 08:02"},
    {"patient_id": "MRN-5521", "sex": "M",      "age": 71,  "analyte": "potassium", "value": 6100.0, "unit": "µEq/L", "collected": "2026-07-01 08:05"},
    {"patient_id": "MRN-1042", "sex": "F",      "age": 62,  "analyte": "potassium", "value": 6.1,    "unit": "mEq/L", "collected": "2026-07-01 07:15"},
    {"patient_id": "MRN-6689", "sex": "m",      "age": 58,  "analyte": "potassium", "value": 3.9,    "unit": "mEq/L", "collected": "07/01/2026 8:40 AM"},
])

# CHECK 1 + 2 — shape and types:
print(labs.shape)
print(labs.dtypes)     # 'value' is object (mixed text/number) — defect found already
```

```python
# CHECK 3 — missing values, per column AND as a fraction:
print(labs.isna().sum())
print((labs.isna().mean() * 100).round(1), "% missing")

# Ask WHY it's missing before deciding what to do:
#  - not measured?  - lost in transfer?  - zero recorded as blank?
# The reason decides the fix. NEVER silently drop without noting it.
```

```python
# CHECK 2 fix — convert text->numeric ONCE, at the boundary, keeping evidence:
labs["value_num"] = pd.to_numeric(labs["value"], errors="coerce")   # unparseable -> NaN
newly_missing = labs["value_num"].isna() & labs["value"].notna()
print(f"{newly_missing.sum()} value(s) could not be parsed:")
print(labs.loc[newly_missing, ["patient_id", "value"]])
```

```python
# CHECK 4 — range audit against DOMAIN limits (not just statistics):
PLAUSIBLE = {"age": (0, 120), "potassium": (1.5, 10.0)}   # wide plausibility, not normal range!

bad_age = labs[(labs["age"] < PLAUSIBLE["age"][0]) | (labs["age"] > PLAUSIBLE["age"][1])]
print("impossible ages:\n", bad_age[["patient_id", "age"]])

lo, hi = PLAUSIBLE["potassium"]
bad_k = labs[(labs["value_num"] < lo) | (labs["value_num"] > hi)]
print("impossible potassium:\n", bad_k[["patient_id", "value_num", "unit"]])
# 6100 µEq/L is 6.1 mEq/L — a UNIT defect, not a lab disaster. Convert, don't delete.
```

```python
# CHECK 5 — duplicates (exact, and same patient+analyte+time):
print("exact duplicate rows:", labs.duplicated().sum())
key_dupes = labs.duplicated(subset=["patient_id", "analyte", "collected"], keep=False)
print(labs.loc[key_dupes, ["patient_id", "collected", "value"]])
```

```python
# CHECK 6 — inconsistent categories and units:
print(labs["sex"].unique())    # ['F' 'female' 'M' 'm'] — FOUR spellings, TWO meanings
print(labs["unit"].unique())   # mixed mEq/L and µEq/L

# Standardize with an explicit mapping (auditable — no magic):
labs["sex_std"] = labs["sex"].str.upper().str[0].map({"F": "F", "M": "M"})
print(labs["sex_std"].unique())

# Unit harmonization: µEq/L -> mEq/L
is_micro = labs["unit"] == "µEq/L"
labs.loc[is_micro, "value_num"] = labs.loc[is_micro, "value_num"] / 1000
labs.loc[is_micro, "unit"] = "mEq/L"
print(labs.loc[is_micro | (labs['patient_id']=='MRN-5521'), ['patient_id','value_num','unit']])
```

```python
# Mixed date formats — parse what you can, FLAG what you can't:
parsed = pd.to_datetime(labs["collected"], format="%Y-%m-%d %H:%M", errors="coerce")
fallback = pd.to_datetime(labs["collected"], format="%m/%d/%Y %I:%M %p", errors="coerce")
labs["collected_dt"] = parsed.fillna(fallback)
print(labs[["patient_id", "collected", "collected_dt"]])
```

### 15c · Fix vs flag — the decision that separates pros from amateurs

| Defect | Default action | Why |
|---|---|---|
| Text-as-number | **Fix** (convert at boundary) | Deterministic, reversible |
| Unit mix-up with clear conversion | **Fix + log** | Value is recoverable |
| Impossible value, cause unknown | **Flag, exclude, investigate** | Guessing fabricates data |
| Missing value | **Depends on WHY** | Week 5: imputation done right |
| Duplicate row | **Fix (drop) + log count** | But same patient re-tested ≠ duplicate! |
| Inconsistent categories | **Fix with explicit mapping** | Mapping is auditable |

**Iron rules:** never edit raw data in place — derive new columns/copies; log every change (what, how many rows, why); the audit trail IS the deliverable. In pharma you already know this instinct: it's ALCOA — attributable, legible, contemporaneous, original, accurate.


```python
# The wrap-up: a reusable QC report function — YOUR first data-quality tool
def qc_report(df, numeric_limits=None):
    """Print the six-check QC summary for any DataFrame."""
    print(f"shape: {df.shape[0]} rows x {df.shape[1]} cols")
    print("\n-- types --");     print(df.dtypes.to_string())
    print("\n-- missing --");   print(df.isna().sum().to_string())
    print(f"\n-- duplicates -- exact: {df.duplicated().sum()}")
    for col in df.select_dtypes("object"):
        u = df[col].unique()
        if len(u) <= 10: print(f"-- categories [{col}]: {list(u)}")
    if numeric_limits:
        for col, (lo, hi) in numeric_limits.items():
            if col in df:
                n = ((df[col] < lo) | (df[col] > hi)).sum()
                print(f"-- range [{col}] outside {lo}-{hi}: {n} row(s)")

qc_report(labs, {"age": (0, 120), "value_num": (1.5, 10.0)})
```

## 16 · Linear algebra in 15 minutes — vectors & matrices with numpy

ML is linear algebra wearing a lab coat. You need exactly three ideas to start:
- a **vector** = a list of numbers = one patient's features
- a **matrix** = a grid = many patients × their features
- the **dot product** = multiply pairwise, then sum = the engine inside every model you'll train


```python
import numpy as np

# A VECTOR: one patient as numbers  [age, weight_kg, creatinine, potassium]
patient = np.array([62, 70, 1.1, 6.1])
print(patient)
print(patient.shape)      # (4,)  -> 4 features
```

```python
# Vector arithmetic is ELEMENTWISE — no loop needed:
weights_kg = np.array([70, 55, 92, 61])
weights_lb = weights_kg * 2.20462      # every element at once
print(weights_lb.round(1))

potassium = np.array([5.2, 4.8, 6.1, 3.2])
print(potassium > 5.0)                 # elementwise comparison -> boolean vector
print((potassium > 5.0).sum())         # True counts as 1 -> how many abnormal
```

```python
# THE DOT PRODUCT: multiply matching entries, add them up.
# This is literally how a model computes a risk score:
features = np.array([62,   70,  1.1, 6.1])    # age, weight, SCr, K
weights  = np.array([0.02, 0.01, 1.5, 0.8])   # "importance" the model LEARNED

risk_score = np.dot(features, weights)
print(risk_score)     # 62*0.02 + 70*0.01 + 1.1*1.5 + 6.1*0.8 = ...

# When we say a model "learns", we mean: it finds good values for `weights`.
# You now understand more of ML than most headlines do.
```

```python
# A MATRIX: rows = patients, columns = features
X = np.array([
    [62, 70, 1.1, 6.1],
    [80, 55, 1.4, 4.8],
    [45, 92, 0.9, 3.2],
])
print(X.shape)          # (3, 4): 3 patients, 4 features
print(X[0])             # first patient (a row)
print(X[:, 3])          # ALL patients' potassium (a column)
print(X.mean(axis=0).round(2))   # mean of each column — instant profiling
```

```python
# Matrix @ vector = every patient's risk score in ONE line (no loop):
scores = X @ weights          # '@' is matrix multiplication
print(scores.round(2))        # three dot products at once
# This single line is what a neural network layer does. Really.
```

**Take away three sentences:** A patient is a vector. A dataset is a matrix. A model's prediction is dot products between your data and learned weights. Weeks 4–16 unpack everything else.

### 18b · Choosing the right chart — a decision guide

| Your question | Chart | Call |
|---|---|---|

### 18c · Chart crimes — what NOT to do (misleading-graph immunity)

1. **Truncated y-axis on bar charts** — bars encode length; starting at 47 instead of 0 turns a 2% difference into a cliff. (Lines may zoom; bars may not.)
2. **Pie charts with >3 slices** — humans can't compare angles. Use a sorted bar.
3. **No units on axes** — "Potassium" of what? 6.1 mEq/L and 6.1 mmol/L happen to match; creatinine µmol/L vs mg/dL differ 88-fold.
4. **Overplotting** — 5,000 scatter dots = one blob. Use `alpha=0.1` or a 2D histogram.
5. **3D anything** for 2D data — perspective distorts every comparison.
6. **Dual y-axes** — two unrelated scales on one plot invite fake correlations.

**The five-second rule:** a stranger should get the point of your chart in five seconds — title says the finding, axes carry units, reference lines mark what matters. You'll be graded on this in every lab report.

## 20 · Power tools — the idioms you'll see everywhere

Eight small tools that appear constantly in real pandas/ML code. None is hard; all are everywhere. Skim now, return when you meet them in the wild.


```python
# 20.1 LIST COMPREHENSIONS — build a list from a loop, in one line
panel = [
    {"patient_id": "MRN-1042", "value": 6.1},
    {"patient_id": "MRN-2117", "value": 4.2},
    {"patient_id": "MRN-3306", "value": 5.4},
]

# loop version                      # comprehension version (same result)
values = []
for r in panel:
    values.append(r["value"])
values2 = [r["value"] for r in panel]
print(values, values2)

# With a condition — filter and extract in one line:
high_ids = [r["patient_id"] for r in panel if r["value"] > 5.0]
print(high_ids)
```

```python
# 20.2 LAMBDA — a tiny unnamed function, used where a function is an ARGUMENT
# Rule of thumb: if it fits on one line and you use it once, lambda; otherwise def.

# Where you'll actually meet it #1: sorting by a key
by_value = sorted(panel, key=lambda r: r["value"], reverse=True)
print([r["patient_id"] for r in by_value])     # highest potassium first

# Where you'll actually meet it #2: pandas .apply()
import pandas as pd
df = pd.DataFrame(panel)
df["status"] = df["value"].apply(lambda v: "HIGH" if v > 5.0 else "ok")
df
```

```python
# 20.3 ENUMERATE & ZIP
meds = ["metformin", "lisinopril", "aspirin"]
doses = ["500 mg", "10 mg", "81 mg"]

for i, med in enumerate(meds, start=1):     # position + item together
    print(f"{i}. {med}")

for med, dose in zip(meds, doses):          # walk two lists in lockstep
    print(f"{med}: {dose}")
```

```python
# 20.4 dict.get() — safe lookups when a key might be missing (data quality!)
rec = {"patient_id": "MRN-1042", "value": 6.1}   # no "unit" key

# rec["unit"] would CRASH (KeyError). .get() returns a default instead:
print(rec.get("unit"))              # None — Python's 'nothing' value
print(rec.get("unit", "UNKNOWN"))   # or your chosen fallback

# Check for None explicitly — the standard missing-data guard:
unit = rec.get("unit")
if unit is None:
    print("unit missing — flag for data quality review")
```

```python
# 20.5 try/except — survive bad data without crashing the whole run
raw_values = ["6.1", "4.2", "N/A", "5.4", ""]

clean, bad = [], []
for v in raw_values:
    try:
        clean.append(float(v))
    except ValueError:
        bad.append(v)

print("clean:", clean)
print("could not parse:", bad)   # logged, not fatal — a 10,000-row file WILL have these
```

```python
# 20.6 while — repeat until a condition changes (loops without a fixed count)
# Example: titrate a dose down until projected level is in range
level = 14.0
dose = 300
while level > 10.0:
    dose -= 25
    level -= 0.9
    print(f"dose {dose} mg -> projected level {level:.1f}")
print("target reached")
# Danger: if the condition can never become False, the loop runs forever.
# (In Colab: Runtime ▸ Interrupt execution rescues you.)
```

```python
# 20.7 SETS — unique values, one line
patient_ids = ["MRN-1042", "MRN-2117", "MRN-1042", "MRN-3306", "MRN-2117"]
unique = set(patient_ids)
print(unique, "-", len(unique), "unique patients from", len(patient_ids), "rows")

# Membership testing is instant, and set math is free:
cohort_a = {"MRN-1042", "MRN-2117"}
cohort_b = {"MRN-2117", "MRN-3306"}
print(cohort_a & cohort_b)    # in both (intersection)
print(cohort_a | cohort_b)    # in either (union)
```

```python
# 20.8 RANDOM + SEEDS — reproducibility, the ML super-habit
import random
random.seed(42)                      # same seed -> same "random" numbers, every run
print(random.random())
print(random.choice(["A", "B", "C"]))

import numpy as np
np.random.seed(42)                   # numpy has its own seed
print(np.random.normal(1.0, 0.25, 3).round(2))

# Every ML experiment you run this semester sets a seed FIRST —
# so results can be reproduced by you, your instructor, and reviewers.
```
