## Worked example — Tokenization
**Lane A · the general idea:** Tokenization is the process of breaking a sentence or document into smaller pieces such as words, subwords, or tokens. The model cannot work directly on raw characters or paragraphs without this step.

**Lane B · the same idea in pharmacy terms:** Clinical notes contain abbreviations, shorthand, and structured fragments. Tokenization is what lets a text model turn those notes into a usable sequence for later classification.

Below is a tiny, complete demo of this idea. Run it, read every line, change one thing, run it again.


```python
# WORKED EXAMPLE — tokenization: text becomes countable pieces
note = "Patient reports nausea after metformin. Metformin dose reduced; nausea resolved."

tokens = note.lower().replace(".", " ").replace(";", " ").split()
print("first tokens:", tokens[:8])

counts = {}
for t in tokens:
    counts[t] = counts.get(t, 0) + 1
print("mentions of 'metformin':", counts["metformin"])
print("mentions of 'nausea':  ", counts["nausea"])
# every NLP model this course touches starts exactly here: text → tokens → numbers.
```
